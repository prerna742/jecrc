```python
tips['day'].value_counts()
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='smoker',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

```python
sns.countplot(x='day',data=tips, hue='sex')
```

```python
sns.barplot(x='day',y='total_bill',data = tips)
```

```python
sns.barplot(x='day',y='total_bill',data = tips,estimator=sum)
```

```python
sns.set_style('whitegrid')
sns.barplot(x='day',y='total_bill',data = tips,estimator=sum)
```

```python
sns.barplot(x='day',y='total_bill',data = tips,hue='time',estimator=np.mean)
```

```python
sns.boxplot(x='day',y='total_bill',data = tips)
```

```python
sns.violinplot(x='day',y='total_bill',data = tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data = tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data = tips,hue='smoker');plt.show()
sns.boxplot(x='day',y='total_bill',data = tips,hue='smoker');plt.show()
sns.violinplot(x='day',y='total_bill',data = tips,hue='smoker');plt.show()
```
