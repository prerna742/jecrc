```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.arange(1,11,3)
```

```python
y=x**3
plt.bar(x,y)
```

```python
plt.bar(x,y)
plt.title('bar graph')
plt.show()
```

```python
plt.hist(x)
plt.title('Histogram')
plt.show()



plt.scatter(x,y)
plt.title('scatter')
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
y2 = x**4
plt.plot(x,y,x,y2)
plt.legend(['y=**3','y2=x**4'])
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.legend(['y=**3','y2=x**4'])
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(x,y,lw=3,marker='o',markersize=15,markerfacecolor='yellow')
plt.plot(x,y2,marker = 'v',markersize=10)
plt.legend(['y=**3','y2=x**4'])
plt.grid()
plt.show()
```

```python
fig.savefig('mygraph.png')
```

```python
plt.boxplot(x)
plt.title('Box plot')
plt.show()
```

```python
plt.stem(x,y)
plt.title('stem')
plt.show()
```

```python
plt.violinplot(x)
plt.title('violin plot - badsoorat wala')
plt.show()
```
